```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
"Modificar Imagen"
fname = '/home/juanmanuel/PycharmProjects/Faces/bush.png' # Ruta de la imagen.
image = Image.open(fname).convert("L") # Se abre la imagen y se aplica una transformacion a blanco y negro.
image.thumbnail((200, 200), Image.ANTIALIAS) # Se cambia el tamaño de la imagen.
image.save('/home/juanmanuel/PycharmProjects/Faces/bush1.png')# se guarda la nueva imagen.

"Desplegar Imagen"
array = np.asarray(image)# Se convierte la imagen a un array de tipo numpy.
plt.imshow(array, cmap='Greys_r')# Se muestra la imagen en los ejes.
plt.show()# Muestra la figura
```
